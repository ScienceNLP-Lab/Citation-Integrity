```python
import os
```

```python
CODES = [
         'AccurateSeg', 
         'MajorSeg', 
         'MinorSeg', 
         'AccurateSection', 
         'MajorSection', 
         'MinorSection', 
         'AccurateEnt', 
         'MajorEnt',
         'MinorEnt'
        ]
```

```python
present_codes = {}

for filename in  os.listdir('AccuracyNew'):
    if '.' not in filename:
        for file in os.listdir(f"AccuracyNew/{filename}"):
            if file.endswith('ann'):
                with open(os.path.join('AccuracyNew', filename, file)) as f:
                    content = f.readlines()
                    a_index = 0
                    new_lines_to_add = []
                    old_lines_to_add = []
                    for c in content:
                        l = c.split('\t')
                        if l[1].split()[0] in CODES:
#                             print(l[1].split()[0])
                            curr_code = l[1].split()[0]
                            new_code = None
                            if curr_code.startswith("Accurate"):
                                new_code = "ACCURATE"
                            elif curr_code.startswith("Minor"):
                                new_code = "Minor"
                            elif curr_code.startswith("Major"):
                                new_code = "Major"
                            
                            new_part = l[1].split()
                            new_part[0] = new_code
                            old_lines_to_add.append("\t".join([l[0], " ".join(new_part), l[2]]))
      
                                
                            if curr_code.endswith("Ent"):
                                s = " ".join(["Level", l[0], 'Article'])
                                nl = '\t'.join([f"A{a_index}", s])
                                a_index += 1
                                new_lines_to_add.append(nl)

                            elif curr_code.endswith("Section"):
                                s = " ".join(["Level", l[0], 'Section'])
                                nl = '\t'.join([f"A{a_index}", s])
                                a_index += 1
                                new_lines_to_add.append(nl)
                        else:
                            old_lines_to_add.append(c)

                                
                    with open(os.path.join('AccuracyNew', filename, file), 'w') as f:
                        f.writelines(old_lines_to_add)
                        f.writelines(new_lines_to_add)
                                

                            
                            
                            
```

```python
res = {}

for filename in os.listdir('AccuracyNew'):
    if '.' not in filename:
        res[filename] = {}
        for file in os.listdir(f"AccuracyNew/{filename}"):
            if file.endswith('ann'):
                res[filename][file[:-4]] = {}
                with open(os.path.join("AccuracyNew", filename, file[:-4]+'.txt')) as f:
                    content = f.read()
                    idx = content.rfind('******')
                    start_idx = idx + 9 # 6 starts above + 3 newline 
      
                with open(os.path.join('AccuracyNew', filename, file)) as f:
                    content = f.readlines()
                   
                    line = content[1].split('\t')[2]
                    res[filename][file[:-4]]['context'] = line
                    res[filename][file[:-4]]['support'] = []
                    

                    for c in content:
                        l = c.split('\t')
                        
                        if l[1].split()[0] in CODES:
                            start, end = int(l[1].split()[1]), int(l[1].split()[2])
                            start, end = start - start_idx, end - start_idx
                       
                            res[filename][file[:-4]]['support'].append([start, end])
```

```python
count = 0
for x in res.values():
    for y in x.values():
        if len(y['support']) > 0:
            count += 1
```
