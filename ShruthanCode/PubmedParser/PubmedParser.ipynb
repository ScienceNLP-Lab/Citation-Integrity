```python
text = ""
text = text + d1["full_title"] + "\n\n"
text = text + "Abstract" + "\n\n" + d1['abstract'] + "\n\n"

curr_section = None

sections = {}
i = 0
for d in dicts_out:
    curr_sec = d['section']
    if curr_sec in sections:
        sections[curr_sec][1] = sections[curr_sec][1] + d['text'] + '\n\n'
    else:
        sections[curr_sec] = [i, ""]
        i += 1
        sections[curr_sec][1] = sections[curr_sec][1] + d['text'] + '\n\n'
        # curr_section = d['section']

sections_list = list(sections.items())
sections_list.sort(key=lambda x : x[1][0])

for sec_name, [_, sec_text] in sections_list:
    text += sec_name + "\n\n" + sec_text 
    
if 'coi_statement' in d1:
    text += d1['coi_statement']
```
